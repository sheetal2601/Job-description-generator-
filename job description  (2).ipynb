```python
import gradio as gr
from transformers import GPT2LMHeadModel, GPT2Tokenizer

# Load pre-trained GPT-2 model and tokenizer
model_name = "gpt2"
model = GPT2LMHeadModel.from_pretrained(model_name)
tokenizer = GPT2Tokenizer.from_pretrained(model_name)
```

```python
def generate_job_description(job_role):
    job_prompts = {
        "Data Scientist": "analyzing large datasets, developing predictive models, and extracting actionable insights",
        "Data Analyst": "analyzing data, generating reports, and providing insights to support business decisions",
        "Software Developer": "designing, developing, and maintaining software solutions",
        "Sales Person": "developing and executing sales strategies, building client relationships, and achieving revenue targets"
    }
    
    if job_role in job_prompts:
        prompt = (
            f"job description for the position of {job_role} at [company name]. "
            f"The {job_role} will be responsible for {job_prompts[job_role]}. "
            f"The ideal candidate should have Relevent skills required. "
        
        )
       
        input_ids = tokenizer.encode(prompt, return_tensors="pt", max_length=400)
        output = model.generate(input_ids, max_length=400, num_return_sequences=1, no_repeat_ngram_size=2)
        generated_text = tokenizer.decode(output[0], skip_special_tokens=True)
       
        return generated_text
    else:
        return "Job role not recognized."
```

```python
inputs = [
    gr.inputs.Textbox(label="Enter Job Title (e.g., data analyst, data scientist, software developer, sales)"),
]

```

```python
from gradio import components
```

```python
output = components.Textbox(label="Generated Job Description")
```

```python
iface = gr.Interface(
    fn=generate_job_description,
    inputs=inputs,
    outputs=output,
    title="Job Description Generator",
    description="Generate job descriptions for specific job titles.",
    examples=[
        ["Data Scientist"],
        ["Data Analyst"],
        ["Software Developer"],
        ["Sales Person"]
    ]
)
```

```python
iface.launch()
```
